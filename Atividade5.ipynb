# Questão 1
Escreva uma função que verifica se um número é par. Se for, imprima uma mensagem. Se não for, imprima outra mensagem.

```python
def verificar_par(numero):
    if numero % 2 == 0:
        print(f"O número {numero} é par.")
    else:
        print(f"O número {numero} é ímpar.")

# Exemplos de uso da função:
verificar_par(10)  # Saída: O número 10 é par.
verificar_par(7)   # Saída: O número 7 é ímpar.
```

# Questão 2
Escreva uma função que calcule o raio de um círculo, dado um valor de raio. Peça um valor de raio ao usuário e calcule o valor da função. (Dica: Use o pacote ``math`` para conseguir o valor de pi)

```python
import math

def calcular_area_circulo(raio):
    # Calcula a área usando math.pi e eleva o raio ao quadrado (** 2)
    area = math.pi * (raio ** 2)
    return area

# Pede um valor de raio ao usuário
entrada_usuario = input("Digite o valor do raio do círculo: ")
raio = float(entrada_usuario)

# Chama a função passando o raio digitado
resultado = calcular_area_circulo(raio)

print(f"A área do círculo com raio {raio} é {resultado:.2f}")
```

# Questão 3 🏆 Desafio Final
 (Algoritmo de ordenação: Bubble sort). Bubble sort é um algoritmo muito simples. Ele começa no início da lista e compara os dois primeiros elementos. Se o primeiro for maior do que o segundo, eles são trocados. Esse processo se repete para cada par subsequente, até o final da lista. Depois o algoritmo volta ao início da lista e recomeça o processo, repetindo-o até não haver mais trocas. Implemente uma função que recebe uma lista e retorna seus elementos ordenados usando o algoritmo bubble sort.
 
 Use a seguinte lista para testar ``l = [10, 1, 2, 8, 0, 9, -1, 11, 5, 4]``
 
 (Dica: Você vai precisar de mais de um tipo de loop nesse exercício)

```python
def bubble_sort(lista):
    # Fazemos uma cópia para não alterar a lista original diretamente
    lista_ordenada = lista.copy() 
    tamanho = len(lista_ordenada)
    houve_troca = True
    
    # O laço while continua rodando enquanto houver trocas na rodada anterior
    while houve_troca:
        houve_troca = False
        
        # O laço for percorre os elementos comparando pares vizinhos
        for i in range(tamanho - 1):
            if lista_ordenada[i] > lista_ordenada[i + 1]:
                # Se o primeiro for maior, trocamos os dois de lugar
                lista_ordenada[i], lista_ordenada[i + 1] = lista_ordenada[i + 1], lista_ordenada[i]
                
                # Registramos que uma troca aconteceu, então o while vai rodar de novo
                houve_troca = True
                
        # Otimização: a cada rodada completa, o maior número sempre "flutua" 
        # para a última posição. Então, não precisamos checar o final novamente.
        tamanho -= 1 
        
    return lista_ordenada

# Testando com a lista fornecida
l = [10, 1, 2, 8, 0, 9, -1, 11, 5, 4]
resultado = bubble_sort(l)

print(f"Lista original: {l}")
print(f"Lista ordenada: {resultado}")
```
